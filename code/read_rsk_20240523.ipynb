```python
from pyrsktools import RSK
from matplotlib import pyplot as plt
```

```python
# plot temperatures

rsk = RSK("../data/SHS/233860_20250213_1924.rsk")
rsk.open()
rsk.readdata()
# rsk.plotdata()
print(rsk.channelNames)

ax = plt.subplot(2,1,1)
ax.plot(rsk.data["timestamp"], rsk.data["temperature"], linewidth=0.3)
ax.plot(rsk.data["timestamp"], rsk.data["temperature1"], linewidth=0.3)
ax.set_ylabel("temperature,\ntemperature1 (°C)")

ax = plt.subplot(2,1,2)
ax.plot(rsk.data["timestamp"], rsk.data["temperature"]-rsk.data["temperature1"], linewidth=0.3)
ax.set_ylabel("temperature-temperature1\n(°C)")
ax.set_ylim([-0.001, 0.005])
ax.plot(rsk.data["timestamp"], 0*rsk.data["temperature"])
ax.plot(rsk.data["timestamp"], 0.0016+0*rsk.data["temperature"])
# rsk.RSK2CSV()

rsk.close()
```

`temperature` - `temperature1` = 0.0016

```python
# convert RSK data to CSV
with RSK("../data/SHS/233860_20250213_1924.rsk") as rsk:
    rsk.open()
    rsk.readdata()
    rsk.RSK2CSV()
    rsk.close()
```
